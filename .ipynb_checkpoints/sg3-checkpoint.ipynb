```python
def make_transform(translate: Tuple[float,float], angle: float):
    m = np.eye(3)
    s = np.sin(angle/360.0*np.pi*2)
    c = np.cos(angle/360.0*np.pi*2)
    m[0][0] = c
    m[0][1] = s
    m[0][2] = translate[0]
    m[1][0] = -s
    m[1][1] = c
    m[1][2] = translate[1]
    return m
```

```python
def num_range(s: str) -> List[int]:
    '''Accept either a comma separated list of numbers 'a,b,c' or a range 'a-c' and return as a list of ints.'''

    range_re = re.compile(r'^(\d+)-(\d+)$')
    m = range_re.match(s)
    if m:
        return list(range(int(m.group(1)), int(m.group(2))+1))
    vals = s.split(',')
    return [int(x) for x in vals]
```

```python
# Generate images from z space.
def gen_image_from_z(z, c=class_idx, label=None):

    # z = torch.from_numpy(np.random.RandomState(seed).randn(1, G.z_dim)).to(device)
    # z: torch tensor of size [1,512]
    
    # Construct an inverse rotation/translation matrix and pass to the generator.  The
    # generator expects this matrix as an inverse to avoid potentially failing numerical
    # operations in the network.
    if hasattr(G.synthesis, 'input'):
        m = make_transform(translate, rotate)
        m = np.linalg.inv(m)
        G.synthesis.input.transform.copy_(torch.from_numpy(m))

    img = G(z, label, truncation_psi=truncation_psi, noise_mode=noise_mode)

    # GRAY
    img = (img + 1) * (255/2)
    img = img.clamp(0, 255).to(torch.uint8)[0].cpu().numpy()
    return PIL.Image.fromarray(img[0], 'L')

    ## COLOR
    #img = (img.permute(0, 2, 3, 1) * 127.5 + 128).clamp(0, 255).to(torch.uint8)
    #return PIL.Image.fromarray(img[0].cpu().numpy(), 'RGB')
```

```python
# Generate images from w space.
def gen_image_from_w(w, c=class_idx):
    
    # w = G.mapping(z, c, truncation_psi=truncation_psi)
    # w: torch tensor of size [1,16,512]

    img = G.synthesis(w, noise_mode=noise_mode, force_fp32=True)
    
    # GRAY
    img = (img + 1) * (255/2)
    img = img.clamp(0, 255).to(torch.uint8)[0].cpu().numpy()
    return PIL.Image.fromarray(img[0], 'L')

    ## COLOR
    #img = (img.permute(0, 2, 3, 1) * 127.5 + 128).clamp(0, 255).to(torch.uint8)
    #return PIL.Image.fromarray(img[0].cpu().numpy(), 'RGB')
```

```python
def get_diff_alongAxis(axis=0):
    
    # Get 3 random samples from w_samples
    samples = [w_samples[i] for i in random.sample(range(len(w_samples)), k=3)]

    img_pre = gen_image_from_w(samples[0].reshape((1,16,512)))
#     img_pre
    
#     print(samples[0])
    samples[0][axis] = (samples[1]+samples[2])[axis]
#     print(samples[0])
    
    img_post = gen_image_from_w(samples[0].reshape((1,16,512)))
    img_diff = ImageChops.difference(img_pre, img_post)
    return np.array(img_diff)
```

```python
row_seeds = num_range('0-9')
col_seeds = num_range('10-19')
```

```python
# row_seeds, col_seeds
```

```python
all_seeds = list(set(row_seeds + col_seeds))
all_seeds
```

```python
all_z = np.stack([np.random.RandomState(seed).randn(G.z_dim) for seed in all_seeds])
all_w = G.mapping(torch.from_numpy(all_z).to(device), None)
w_avg = G.mapping.w_avg
all_w = w_avg + (all_w - w_avg) * truncation_psi
w_dict = {seed: w for seed, w in zip(all_seeds, list(all_w))}
all_images = G.synthesis(all_w, noise_mode=noise_mode)
all_images = ((all_images+1)*(225/2)).clamp(0, 255).to(torch.uint8).cpu().numpy()
image_dict = {(seed, seed): image for seed, image in zip(all_seeds, list(all_images))}
W = G.img_resolution
H = G.img_resolution
outdir = './'
```

```python
from skimage.metrics import structural_similarity as ssim
```

```python
def save_canvas(nr):
    col_styles = num_range(nr)
    for row_seed in row_seeds:
        for col_seed in col_seeds:
            
            w = w_dict[row_seed].clone()
#             image_left = G.synthesis(w[np.newaxis], noise_mode=noise_mode)
#             image_left = ((image_left+1)*127.5).clamp(0, 255).to(torch.uint8)
    
            w[col_styles] = w_dict[col_seed][col_styles]
            image_new = G.synthesis(w[np.newaxis], noise_mode=noise_mode)
            image_new = ((image_new+1)*127.5).clamp(0, 255).to(torch.uint8)
                
            image_dict[(row_seed, col_seed)] = image_new[0].cpu().numpy()

                
    canvas = PIL.Image.new('L', (W * (len(col_seeds) + 1), H * (len(row_seeds) + 1)), 'black')

    for row_idx, row_seed in enumerate([0] + row_seeds):
        for col_idx, col_seed in enumerate([0] + col_seeds):
            if row_idx == 0 and col_idx == 0:
                continue
            key = (row_seed, col_seed)
            if row_idx == 0:
                key = (col_seed, col_seed)
            if col_idx == 0:
                key = (row_seed, row_seed)
            canvas.paste(PIL.Image.fromarray(image_dict[key][0], 'L'), (W * col_idx, H * row_idx))
            
    canvas.save(f'{outdir}/nr-{nr}.png')
```

```python
for nr in nrs:
    print(nr)
    save_canvas(nr)
```
